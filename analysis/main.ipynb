```python
import json
import pandas as pd
import utils as u
import matplotlib.pyplot as plt
```

```python
with open("creds.json") as file:
    creds = json.load(file)
```

```python
query = '''
select visitor_id, visitor, home_id, home, Vpoints, Hpoints, mov, home_victory, year(date) year from games where year(date) = 2022 order by year(date)
'''
data = u.sqlTodf(query, creds)
data
```

Clear home court advantage

```python
n = len(data.index)
homeWin = len(data[data["home_victory"] == 1]) * 100 / n
homeLoss = len(data[data["home_victory"] == 0]) * 100 / n
homeWin, homeLoss
```

```python
teamWL = dict()
for _, row in data.iterrows():
    mov = row["mov"]
    home = row["home"]
    visitor = row["visitor"]
    if home not in teamWL:
        if mov > 0:
            teamWL[home] = [1, 0, 1]
        else:
            teamWL[home] = [0, 1, 1]
    else:
        if mov > 0:
            teamWL[home][0] += 1
        else:
            teamWL[home][1] += 1
        teamWL[home][2] += 1

    if visitor not in teamWL:
        if mov > 0:
            teamWL[visitor] = [1, 0, 1]
        else:
            teamWL[visitor] = [0, 1, 1]
    else:
        if mov > 0:
            teamWL[visitor][0] += 1
        else:
            teamWL[visitor][1] += 1
        teamWL[visitor][2] += 1
teamWL
    
```
