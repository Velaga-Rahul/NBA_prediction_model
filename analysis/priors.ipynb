```python
import json
import utils as u
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open("creds.json") as file:
    creds = json.load(file)

creds
```

```python
teamIds = 2, 11

query = '''
select * from games where visitor_id in (2, 11) and home_id in ({}, {})
'''.format(teamIds[0], teamIds[1])
```

```python
data = u.sqlTodf(query, creds)
n = len(data.index)
data
```

Naive Priors

```python
victoryCondition = ((data["home_id"] == teamIds[0]) & (data["home_victory"] == 1)) | ((data["visitor_id"] == teamIds[0]) & (data["home_victory"] == 0))
team0Victory = len(data[victoryCondition].index)
prior0 = team0Victory / n

victoryCondition = ((data["home_id"] == teamIds[1]) & (data["home_victory"] == 1)) | ((data["visitor_id"] == teamIds[1]) & (data["home_victory"] == 0))
team1Victory = len(data[victoryCondition].index)
prior1 = team1Victory / n

priors = prior0, prior1
priors
```
